```python
test = test.fillna(test['prod_cost'].mean)
```

```python
warranty = test[["warranty"]]

print(warranty)
```

```python
test["warranty"] = test["warranty"].str[0].astype(int)
```

```python
test.loc[test['price'] < 300 , 'price'] = 0-300 
```

```python
def cprice(price):
    if price <= 301:
        return '0-300'
    elif (price >= 300) & (price <= 500):
        return '301 - 500'
    elif price >= 501:
        return '500++'
    
test['price'] = test['price'].map(cprice)

```

```python
def ctype(product_type):
    if product_type =='auto-portee' :
        return 1    
    if product_type =='electrique' :
        return 2   
    if product_type =='essence' :
        return 3

test['product_type'] = test['product_type'].map(ctype)
test.groupby(['product_type']).first()
```
