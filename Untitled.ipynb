```python
import pandas as pd
import numpy as np

test= pd.read_csv('mower_market_snapshot.csv', sep = ';')
test
```

```python
print(test['prod_cost'].describe())
print(test['prod_cost'].shape)
```

```python
def null(dataframe):
    return pd.isnull(dataframe)
display(null(test).sum())
```

```python
test = test.fillna(test['prod_cost'].mean)
```

```python
test.warranty.unique()
```

```python
warranty = test[["warranty"]]

print(warranty)
```

```python
test["warranty"] = test["warranty"].str[0].astype(int)
```

```python
test.loc[test['price'] < 300 , 'price'] = 0-300 
```

```python
def cprice(price):
    if price <= 301:
        return '0-300'
    elif (price >= 300) & (price <= 500):
        return '301 - 500'
    elif price >= 501:
        return '500++'
    
test['price'] = test['price'].map(cprice)

```

```python
test
```

```python
def ctype(product_type):
    if product_type =='auto-portee' :
        return 1    
    if product_type =='electrique' :
        return 2   
    if product_type =='essence' :
        return 3

test['product_type'] = test['product_type'].map(ctype)
```

```python
test.groupby(['price']).first()
```

```python
test["auto-portee"]=0
test["electrique"]=0
test["essence"]=0

test.loc[test["product_type"]==1,"auto-portee"]=1
test.loc[test["product_type"]==2,"electrique"]=1
test.loc[test["product_type"]==3,"essence"]=1

test.loc[:,['essence' , 'auto-portee' ,'electrique']]
```
